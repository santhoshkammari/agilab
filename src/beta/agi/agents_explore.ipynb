```python
from pathlib import Path

text = Path('/home/ntlpt59/Downloads/tiger.txt').read_text()[:8000]
print(len(text))
```

```python
from dataclasses import dataclass
from tqdm import trange
import uuid

@dataclass
class Node:
    content:str
    id:str = None
    childs:list["Node"]=None

    def __post_init__(self):
        self.id = str(uuid.uuid4())

leafs = [Node(text[x:x+100]) for x in range(0,len(text),1000)]

async def get_summary_node(nodes):
    prompts = [f'summarize to 1000 words \n\n {n.content} \n /no_think' for n in nodes]
    res = await lm.async_batch_call_llm(prompts)
    content = "".join([r.content for r in res])
    return Node(content)

while len(leafs)>1:
    print(len(leafs))
    summaries = []
    for i in trange(0,len(leafs),3):
        res = await get_summary_node(leafs[i:i+5])
        summaries.append(res)
    leafs = summaries

print(len(leafs))
print(leafs[0].content.replace("""<think>

</think>""",""))

```
